```python
from qiskit import QuantumCircuit
```

```python
qc = QuantumCircuit(3)
qc.ccx(2,1,0) #2,1 are control qubit and 0 is the target qubit
print(qc)
```

```python
qc1 = QuantumCircuit(4)
qc1.mcx([1,2,3],0)
print(qc1)
```

```python
qc2 = QuantumCircuit(3)
qc2.cswap(2,1,0) #2 is the control qubit
print(qc2)
```
