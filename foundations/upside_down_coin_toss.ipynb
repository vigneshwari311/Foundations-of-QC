```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

### Here we can change the qubit state to 1 basically there are two ways to do this operation
#### 1) Add x gate at the start which flips 0 --> 1

```python
q_ct1 = QuantumCircuit(1,1)
q_ct1.x(0)
q_ct1.h(0)
q_ct1.measure(0,0)
q_ct1.draw('mpl')
```

```python
sim = AerSimulator()
job = sim.run(q_ct1,shots = 1)
result = job.result()
```

```python
count = result.get_counts()
print(count)
```

```python
display(plot_histogram(count))
```

#### 2) We can change the initial state of the qubit to 1

```python
q_ct2 = QuantumCircuit(1,1)

initialize_state_vector = [0. + 0.j, 1.+0.j]
```

```python
q_ct2.initialize(initialize_state_vector,0)
```

```python
q_ct2.h(0)
q_ct2.measure(0,0)
q_ct2.draw()
```
