```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_histogram
from qiskit_aer import AerSimulator
```

```python
qc = QuantumCircuit(2,2)
qc.h([0,1])
qc.measure([0,1],[0,1])
qc.draw('mpl')
```

```python
sim = AerSimulator()
job = sim.run(qc,shots = 1)
result = job.result()
```

```python
plot_histogram(result.get_counts())
```
