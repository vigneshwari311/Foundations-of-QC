```python
import numpy as np
from math import pi, sin, cos
from qiskit.visualization import plot_bloch_vector
import matplotlib.pyplot as plot
```

```python
angles_1 = {"theta" : pi/2, "phi":0}
```

```python
bits = {
    "bits = 0" : {"theta" : 0, "phi":0},
    "bits = 1" : {"theta" : pi, "phi":0},
    "|0\u27E9" : {"theta" : 0, "phi":0},
    "|1\u27E9" : {"theta" : pi, "phi":0},
    "a|0⟩+b|1⟩": angles_1
}
```

```python
for label in bits:
    theta = bits[label]["theta"]
    phi = bits[label]["phi"]
    
    x = cos(phi)*sin(theta)
    y = sin(theta)*sin(phi)
    z = cos(theta)
    
    bloch_vector = [x,y,z]
    
    print(f"{label} the vector is {bloch_vector}")
    
    display(plot_bloch_vector(bloch_vector,title=label))

plot.show()
```
