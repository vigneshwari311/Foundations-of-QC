```python

from qiskit import QuantumCircuit
from qiskit.visualization import plot_bloch_multivector, plot_state_qsphere
from qiskit.quantum_info import Statevector,Operator
from PIL import Image
import os
```

```python
dirName = 'images'
if not os.path.exists(dirName):
    os.mkdir(dirName)
    
def get_psi(circuit):
    global psi
    return Statevector.from_instruction(circuit)
     
```

```python
def create_images(gate,theta=0.0,phi=0.0,lam=0.0):
    steps = 20
    theta_steps = theta/steps
    phi_steps = phi/steps
    lam_steps = lam/steps
    
    global q_images,b_images,q_filename,b_filename
    b_images=[]
    q_images=[]
    b_filename="animated_qubit"
    q_filename="animated_qsphere"
    
    for n in range(steps+1):
        
        qc = QuantumCircuit(1)
        
        curr_t = theta_steps * n
        curr_p = phi_steps * n
        curr_l = lam_steps * n
        
        qc = QuantumCircuit(1)
        if gate == "u3":
            qc.u(curr_t, curr_p, curr_l, 0)
            title = f"U3: \u03B8={curr_t:.2f} \u03D5={curr_p:.2f} \u03BB={curr_l:.2f}"
        elif gate == "u2":
            qc.u(1.57, curr_p, curr_l, 0)
            title = f"U2: \u03D5={curr_p:.2f} \u03BB={curr_l:.2f}"
        else: # u1 case
            qc.h(0)
            qc.u(0, 0, curr_p, 0)
            title = f"U1: \u03D5={curr_p:.2f}"
            
        plot_bloch_multivector(get_psi(qc),title).savefig('images/bloch'+str(n)+'.png')
        imb = Image.open('images/bloch'+str(n)+'.png')
        b_images.append(imb)
        
        plot_state_qsphere(get_psi(qc)).savefig('images/qsphere'+str(n)+'.png')
        imq = Image.open('images/qsphere'+str(n)+'.png')
        q_images.append(imq)
        
        n+=1
        theta += theta_steps
        phi += phi_steps
        lam += lam_steps
```

```python
def save_gif(gate):
    duration = 100
    b_images[0].save(f"{gate}_{b_filename}.gif", save_all = True,
                     append_images = b_images[1:],duration = duration,loop=0)
    
    q_images[0].save(f"{gate}_{q_filename}.gif", save_all = True,
                    append_images = q_images[1:],duration = duration,loop=0)
```

```python
def main():
    global gate
    gate = ""
    theta = phi = lam = 0.0
    
   
    gate = input("Enter u1,u2,u3 (or exit): ")
    
    if gate in ["u3","u2","u1"]:
        if gate == "u3":
            theta = float(input("Enter \u03B8:\n"))
        if gate in ["u3","u2","u1"]:
            phi=float(input("Enter \u03D5:\n"))
        if gate in ["u3","u2"]:
            lam = float(input("Enter \u03BB:\n"))
        
        print("Building animation")
        create_images(gate,theta,phi,lam)
        save_gif(gate)
        print(f"Successfully created the animation!!!")
    else:
        print("Not valid input, try again...")
        
if __name__ == '__main__':
    main()
        
            
    
```
