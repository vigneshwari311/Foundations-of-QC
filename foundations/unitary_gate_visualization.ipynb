```python

from qiskit import QuantumCircuit
from qiskit.visualization import plot_bloch_multivector, plot_state_qsphere
from qiskit_aer import Aer
```

```python
def create_images(gate,theta,phi,lam):
    steps = 20.0
    n = 0.0
    theta_steps = theta/steps
    phi_steps = phi/steps
    lam_steps = lam/steps
    
    while n<steps+1:
        qc = QuantumCircuit(1)
        
        if gate == "u3":
            qc.u3(theta,phi,lam,0)
            title = f"U3: \u03B8 = {theta:.2f}  \u03D5 = {phi:.2f}  \u03BB = {lam:.2f}"
        elif gate == "u2":
            qc.u2(phi,lam,0)
            title = f"U2: \u03D5 = {phi:.2f}  \u03BB = {lam:.2f}"
        else:
            qc.h(0)  #by applying h gate we can see the phase, where in default z-axis we cannot see the phase
            qc.u1(phi,0)
            title = f"U1: \u03D5 = {phi:.2f}"
```
